# **BAR GRAPH:**

```
import matplotlib.pyplot as plt
x=pd.read_csv('/content/FL_insurance_sample.csv')
t=x.groupby("construction")[["tiv_2012","tiv_2011"]].mean()
t.plot(kind='bar',color=["green","red"])
plt.title("Comparision between construction and tiv in Florida ")
plt.xlabel("construction")
plt.ylabel("Total insured values(tiv) of years 2011,2012")
plt.show()
```

# **LINE PLOT:**

```
d=x["hu_site_limit"]
u=x['tiv_2011']
t=x['tiv_2012']
plt.plot(d,u,label="tiv_2011",color="blue")
plt.plot(d,t,label="tiv_2012",color="green")
plt.legend()
plt.title("compoarision between hu_site_limit and tiv",size=15,color="red")
plt.xlabel("hu_site_limit")
plt.ylabel("tiv_2011 and tiv_2012")
plt.show()
```

# **BOX PLOT:**

```
import seaborn as sns
e=sns.boxplot(x="construction",y="point_latitude",data=x)
e
```

###**NOTE:**- From the above graph the out liers are said to be zero.

# **REGRESSION PLOT:**

```
import seaborn as sns
e=sns.regplot(x="tiv_2011",y="point_latitude",data=x,color="blue",marker="v")
e
```

# **VIOLIN PLOT:**

```
import seaborn as sns
e=sns.violinplot(x="construction",y="point_latitude",data=x)
e
```

# **SWARM PLOT:**

```
import seaborn as sns
s=sns.swarmplot(x="construction",y="point_latitude",data=x)
s
```

# **HISTOGRAM:**

```
import matplotlib.pyplot as plt
d=x["point_granularity"]
d.plot(kind="hist",bins=10,rwidth=0.8,label="frequency",color="violet")
plt.legend()
plt.xlabel("bins")
plt.title("Histogram to study granularity")
plt.show()
```

## **PIE CHART:**

```

import matplotlib.pyplot as plt
t=x.groupby("construction")["tiv_2011"].mean()
t.plot(kind='pie',explode=[0,0,0,0.3,0],autopct="%1.1f%%",startangle=90)
plt.title("Comparision between construction and tiv in Florida ")
plt.legend()
plt.show()
```

# **STRIP PLOT:**

```
import seaborn as sns
e=sns.stripplot(x="construction",y="point_longitude",data=x,jitter=0.09)
e
```
